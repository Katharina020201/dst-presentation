## Data Cleaning

```python
# dropping not interesting cloumns, e.g. website, founders, ...

df.drop(['Brief Description', 'Animal-Type Analog', 'State', 'City', 'Website', 'Founders'], axis=1, inplace=True)

df.head()
```

```python
# transforming objects to categories 

LIST_CAT = ['Protein Category','Company Focus','Company type','Technology Focus','Product Type','Ingredient Type','Operating Regions','Country/Region']

for i in LIST_CAT:
    df[i] = df[i].astype('category')

df.info()
```

```python
#converting the year from float to int

df['Year Founded'] = df['Year Founded'].fillna(0).astype(int)

df['Year Founded'].head()
```

```python
# converting the company names as strings 

df['Company'] = df['Company'].fillna(0).astype('string')

df.info()
```

```python
# Flitering: Only Protein Category "Plant-based" is relevant for this analysis

df = df[df['Protein Category'] == 'Plant-based']

# drop Logo + Operating regions cloumn, not needed for this analysis
df = df.drop(columns=['Logo'])
# df = df.drop(columns=['Operating Regions'])

df.head()
```

## New Dataframe with companies founded by year

```python
# (Help from GitHub Copilot)
# Grouping by year founded and counting the number of companies, also excluding Conpany Type "Diverisified", beacuse I want only the companies that are specialized in alternative proteins (not Unilever, Nestle, etc. who only have some alternatives)

df_founded = df[df['Company type'] == 'Specialized (focused on alternative proteins)'].groupby('Year Founded').size().reset_index(name='Number of Companies Founded')

df_founded.head()
```

Thoughts on this explorative analysis: Maybe it is not possible to compare the number of companies founded by year, because the data is not complete? The data set is **updated by submission**, so maybe companies founded after 2021 are not that popular yet and therefore not included in the data set?

-> Decission: I will use the data up until 2021, because it fits better with the meat consumption data & fits better in the storyline

```python
# include only companies where year founded is >= 2010 && <= 2021, to compare data with the meat consumption per person data (which starts in 2010)

df_founded = df_founded[(df_founded['Year Founded'] >= 2010) & (df_founded['Year Founded'] <= 2021)]

df_founded.head()
```

```python
df_founded.head(20)
```

## Plot Dataframe with companies founded by year

```python
# Convert year founded to datetime

df_founded['Year Founded'] = pd.to_datetime(df_founded['Year Founded'], format='%Y')

# Plotting line chart

line_chart = alt.Chart(df_founded
).mark_line().encode(
    x=alt.X('Year Founded').axis(
        title = "Year",
        titleColor = 'grey',
        titleAnchor='start',
        labelAngle = 0,
        grid = False,
        tickColor= 'grey'), 
    y= alt.Y('Number of Companies Founded').axis(
        title ='Number of Companies Founded',
        titleColor= 'grey',
        titleAnchor='end',
        grid = False, 
        tickColor= 'grey', 
    ),
    strokeWidth = alt.value(2)
).properties(
    width = 550,
    height = 350
)

line_chart 

```

```python
df_founded.info()
```

```python
# Problem: df_founded['Year Founded'] could not be copared with 2019, because it is a datetime object
# help from Copilot

# create a new column 'Year' that contains just the year part of 'Year Founded'
df_founded['Year_int'] = df_founded['Year Founded'].dt.year

founded_point = alt.Chart(df_founded[df_founded['Year_int'] == 2019]).mark_circle(opacity=1, size=200).encode(
        x=alt.X('Year Founded:T'),
        y=alt.Y('Number of Companies Founded:Q'),
        color=alt.value('red')
)


line_chart + founded_point
```

```python
# label for 2019-point 

label_2019_base = alt.Chart(df_founded[df_founded['Year_int'] == 2019]).mark_text().encode(
    x=alt.X('Year Founded:T'),
    y=alt.Y('Number of Companies Founded:Q'),
    color=alt.value('red'),
    text=alt.Text('Number of Companies Founded:Q')
)

label_2019 = label_2019_base.mark_text(
    align='left',
    dx=25,
    size=20,
)

line_chart + founded_point + label_2019
```

to check if any countries are lost when merging with the country_codes data set in the map visualization, I will compare my country names with the ones in the country_code data

```python
url = 'https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv'
df_country_codes = pd.read_csv(url)

df_country_codes.head()
```

```python
df_country_codes.info()
```

```python
merged_df = df_country.merge(df_country_codes, left_on='Country/Region', right_on='name', how='left')

merged_df
```

now looking at the country-code column, which will be used for the map visualization... There are missing values. Because i will only plot europe, i will edit the names of european countries manually. 

European countries in df_country affected:
- Bavaria and Scotland (can be dropped, only one Company in each)
- Czech Republic -> in country_code file: Czechia
- United Kingdom -> in country_code file: United Kingdom of Great Britain and Northern Ireland

```python
## Help from Copilot add_categories

df_country['Country/Region'] = df_country['Country/Region'].cat.add_categories(['United Kingdom of Great Britain and Northern Ireland', 'Czechia'])

df_country.loc[df_country['Country/Region'] == 'United Kingdom', 'Country/Region'] = 'United Kingdom of Great Britain and Northern Ireland'
df_country.loc[df_country['Country/Region'] == 'Czech Republic', 'Country/Region'] = 'Czechia'


df_country
```

```python
df_country = df_country.rename(columns={"Country/Region": "name"})
```

## Plotting map visulization (with help from source below)

```python
## ----!!! Source: https://github.com/bast/altair-geographic-plots/blob/main/choropleth.ipynb  ---- ##

countries = alt.topo_feature(data.world_110m.url, "countries")
# https://en.wikipedia.org/wiki/ISO_3166-1_numeric
country_codes = pd.read_csv(
    "https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv"
)

background = alt.Chart(countries).mark_geoshape(fill="lightgray")

# we transform twice, first from "ISO 3166-1 numeric" to name, then from name to value
foreground = (
    alt.Chart(countries)
    .mark_geoshape()
    .transform_lookup(
        lookup="id",
        from_=alt.LookupData(data=country_codes, key="country-code", fields=["name"]),
    )
    .transform_lookup(
        lookup="name",
        from_=alt.LookupData(data=df_country, key="name", fields=["Number of Companies per Country"]),
    )
    .encode(
        fill=alt.Color(
            "Number of Companies per Country:Q",
            scale=alt.Scale(scheme="reds"),
            title="Number of Companies"
        ),
        tooltip=[alt.Tooltip('name:N', title='Country'), alt.Tooltip('Number of Companies per Country:Q', title='Number of Companies')]
    )
)

chart = (
    (background + foreground)
    .properties(width=700, height=600)
    .project(
        type="equalEarth",
        scale=800,
        translate=[200,1000],
    )
)

chart
```
